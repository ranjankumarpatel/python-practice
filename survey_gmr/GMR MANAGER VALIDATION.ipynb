```python
sql1 = """SELECT surveyuser0_.DEMOGRAPHIC_NAME AS col_0_0_,
       surveyuser0_.DEMOGRAPHIC_VALUE AS col_1_0_m,count(surveyuser1_.USD_ID) as count_c
  FROM survey_v1_user_demographic_details surveyuser0_
       INNER JOIN survey_v1_user_servey_def surveyuser1_
          ON surveyuser0_.USD_ID = surveyuser1_.USD_ID
       INNER JOIN survey_v1_defination surveydefi2_
          ON surveyuser1_.SURVEY_DEF_ID = surveydefi2_.SD_ID
 WHERE     surveydefi2_.SD_ID = 12
       AND (surveyuser1_.USD_ID IN (SELECT surveyuser5_.USD_ID
                                      FROM survey_manager surveymana3_
                                           INNER JOIN
                                           manager_survey_user_def_mapping managersur4_
                                              ON surveymana3_.MANAGER_ID =
                                                    managersur4_.MANAGER_ID
                                           INNER JOIN
                                           survey_v1_user_servey_def surveyuser5_
                                              ON managersur4_.SUSD_ID =
                                                    surveyuser5_.USD_ID
                                     WHERE surveymana3_.MANAGER_EMP_ID = 6005 and surveyuser5_.SURVEY_DEF_ID = 12))
       AND surveyuser1_.SUBMISSION_STATUS = 'Y'
GROUP BY surveyuser0_.DEMOGRAPHIC_NAME,surveyuser0_.DEMOGRAPHIC_VALUE
  HAVING count(surveyuser1_.USD_ID) > 5
"""
```
